```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame

pw_electric = pd.read_csv(r'PW_electric.csv', sep=',',skiprows=2, header=0,encoding = "utf-8",index_col=0)
pw_electric.head()
```

```python
pw_electric_luzern = DataFrame(pw_electric.ix["Luzern",1:])
pw_electric_luzern
pw_electric_luzern["Year"] = pd.DatetimeIndex(pw_electric_luzern.index)
pw_electric_luzern.set_index("Year", inplace=True)
pw_electric_luzern.plot()
plt.xlabel("Jahr")
plt.ylabel("Anzahl Elektro-Autos Luzern")
plt.show()
```

```python
pw_electric_zuerich = DataFrame(pw_electric.ix["Zürich",1:])
pw_electric_zuerich
pw_electric_zuerich["Year"] = pd.DatetimeIndex(pw_electric_zuerich.index)
pw_electric_zuerich.set_index("Year", inplace=True)
pw_electric_zuerich.plot()
plt.xlabel("Jahr")
plt.ylabel("Anzahl Elektro-Autos Zürich")
plt.show()
```
